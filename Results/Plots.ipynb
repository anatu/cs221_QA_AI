```python
import os, re
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns; sns.set(rc={'figure.figsize':(11.7*2,8.27*2)})
%pylab inline
```

```python
   
mem2_hist = json.load(open("BABI_MEMNN_TWO_HOP_QA2_history.json", "r"))
mem1_hist = json.load(open("BABI_MEMNN_ONE_HOP_QA2_history.json", "r"))
baseline_hist = json.load(open("BASELINE_QA2_history.json", "r"))
```

```python
xlen = len(baseline_hist["val_loss"])
print(xlen)
x_pts = [i for i in range(xlen)]
```

```python
palette = sns.color_palette("husl", 3)
loss_df = pd.DataFrame()
loss_df["MemNN (2-Hop)"] = mem2_hist["val_loss"]
loss_df["MemNN (1-Hop)"] = mem1_hist["val_loss"]
loss_df["Baseline GRU"] = baseline_hist["val_loss"]
accplot = sns.lineplot(data = loss_df, markers = False, palette = palette)
accfig = accplot.get_figure()
accfig.savefig("val_loss.png")
```

```python
palette = sns.color_palette("husl", 3)
acc_df = pd.DataFrame()
acc_df["MemNN (2-Hop)"] = mem2_hist["val_acc"]
acc_df["MemNN (1-Hop)"] = mem1_hist["val_acc"]
acc_df["Baseline GRU"] = baseline_hist["val_acc"]
accplot = sns.lineplot(data = acc_df, markers = False, palette = palette)
accfig = accplot.get_figure()
accfig.savefig("val_acc.png")
```
