```python
import os, re
import json
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%pylab inline
```

```python
mem2_hist = json.load(open("BABI_MEMNN_TWO_HOP_QA2_history.json", "r"))
baseline_hist = json.load(open("BASELINE_QA2_history.json", "r"))
```

```python
xlen = len(baseline_hist["val_loss"])
print(xlen)
x_pts = [i for i in range(xlen)]
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey = False)
fig.set_figheight(4)
fig.set_figwidth(15)

sns.lineplot(x=x_pts, y=baseline_hist["loss"], ax = ax1).set_title("loss")
sns.lineplot(x=x_pts, y=mem2_hist["loss"], ax = ax1).set_title("loss")
sns.lineplot(x=x_pts, y=baseline_hist["acc"], ax = ax2).set_title("acc")
sns.lineplot(x=x_pts, y=mem2_hist["acc"], ax = ax2).set_title("acc")

fig.savefig(os.path.join("train_metrics.png"))
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey = False)
fig.set_figheight(4)
fig.set_figwidth(15)
sns.lineplot(x=x_pts, y=baseline_hist["val_loss"], ax = ax1).set_title("loss")
sns.lineplot(x=x_pts, y=mem2_hist["val_loss"], ax = ax1).set_title("loss")
sns.lineplot(x=x_pts, y=baseline_hist["val_acc"], ax = ax2).set_title("acc")
sns.lineplot(x=x_pts, y=mem2_hist["val_acc"], ax = ax2).set_title("acc")
fig.savefig(os.path.join("val_metrics.png"))
```
