```python
import os, re
import json
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%pylab inline
```

```python
json_path = "Outputs/baseline_history.json"
hist = json.load(open(json_path, "r"))
```

```python
print(hist["loss"])
```

```python
xlen = len(hist[list(hist.keys())[0]])
print(xlen)
x_pts = [i for i in range(xlen)]
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey = False)
fig.set_figheight(4)
fig.set_figwidth(15)
sns.lineplot(x=x_pts, y=hist["loss"], ax = ax1).set_title("loss")
sns.lineplot(x=x_pts, y=hist["acc"], ax = ax2).set_title("acc")
fig.savefig("Outputs/train_metrics.png")
```

```python
fig, (ax1, ax2) = plt.subplots(ncols=2, sharey = False)
fig.set_figheight(4)
fig.set_figwidth(15)
sns.lineplot(x=x_pts, y=hist["val_loss"], ax = ax1).set_title("val_loss")
sns.lineplot(x=x_pts, y=hist["val_acc"], ax = ax2).set_title("val_acc")
fig.savefig("Outputs/val_metrics.png")
```
